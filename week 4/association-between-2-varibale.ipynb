# Association between two random variables

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the housing information for analysis 

housing = pd.read_csv('../data/housing.csv', index_col=0)
housing.head()
```

```python
# Use covariance to calculate the association

housing.cov()
```

```python
# Use correlation to calculate the association is more appropriate in this case
housing.corr()
```

```python
# scatter matrix plot
pd.plotting.scatter_matrix(housing, figsize=(8, 8))
plt.show()
```
